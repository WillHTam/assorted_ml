## Data Setup

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np

np.random.seed(8)
```

```python
data = np.loadtxt('./pima-indians-diabetes.data.csv', delimiter=',')
```

Load first eight columns of 
    1. number times pregnant
    2. glucose concentration
    3. diastolic
    4. tricep skin fold thickness
    5. 2-hour serum insulin
    6. BMI
    7. Diabetes pedigree function
    8. Age

```python
X = data[:, 0:8]
```

Predict diabetes <br/>
0 (negative) - 500 instances <br/>
1 (positive) - 268 instances <br />
From ninth column 

```python
Y = data[:, 8]
```

## Defining Model

Input layer should have the right number of inptus

Therefore `input_dim` should be 8 for the 8 input variables

How many layers should there be? 
- Depends. Usually depends on trial & error, should be large enough to process the structure of the problem

### Three layer neural network
- Create a fully-connected network structure with three layers
    - Fully connected layers use the 'Dense' class
- Use rectifier activation `relu` for the first two layers
    - Better performance with relu
- Use sigmoid activation for the last layer
    - This ensures an output between 0 and 1. 
    - Threshold of 0.5 which separates the decision

```python
model = Sequential()
# 1st layer: 12 neurons, expects 8 inputs
model.add(Dense(12, input_dim=8, activation='relu'))
# 2nd layer: 8 neurons
model.add(Dense(8, activation='relu'))
# 3rd layer: 1 neuron to predict the classs
model.add(Dense(1, activation='sigmoid'))
```

## Compiling Model

In compiling, must define
    1. loss function to evaluate the weights
    2. optimizer to search through different weights for the network
    3. any optional metrics to collect and report during training
Because it is a classification problem, we will collect and report the classification accuracy as the metric.

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

## Fitting Model

Use model.fit() to start training

`epochs` argument is the number of training iterations through the dataset

`batch_size` is the number of instances that are evaluated before a weight update in the network is performed<br/>
_Setting batch size higher makes the process faster on GPU's but also can degrade the model if set excessively high_
    

```python
# 150 is considered a small amount of iterations
# Don't show display bars with verbose=0
model.fit(X, Y, epochs=150, batch_size=10)
```

## Evaluating Model

```python
# evaluate the model
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

## Making predictions

```python
predictions = model.predict(X)
```
